```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv("Question3 Output.csv");
```

```python
data.head()
```

```python
x_axis = np.arange(len(data["date"]))
plt.bar(x_axis -0.2, data["earlyLockDown"], width=0.2, label = 'ELD')
plt.bar(x_axis +0.2, data["lockDownInApril"], width=0.2, label = 'LDA')

plt.xticks(x_axis, data["date"],rotation=90)
plt.gcf().set_size_inches(30, 10)

plt.xlabel("Date",fontsize=20)
plt.ylabel("New Cases",fontsize=20)
plt.title("US- LockDown vs Early Lockdown",fontsize=30)
plt.legend(["ELD", "LD"])
plt.show()
plt.savefig("LDvsELD.png")
```

```python
data.plot(x="date", y=["earlyLockDown","lockDownInApril" ], kind="line", figsize=(10, 7))
plt.xlabel("Date",fontsize=20)
plt.ylabel("New cases",fontsize=20)
plt.title("US- Lock Down vs Early Lock Down",fontsize=30)
plt.show()
plt.savefig("US-DeathCasesLineGraph.png")
```
